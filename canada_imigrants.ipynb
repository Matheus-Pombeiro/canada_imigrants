## **Imigrantes no Canadá**

## Conhecendo a biblioteca Matplotlib

```python
# Importando bibliotecas
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Recebendo a url do dataset
url = 'https://raw.githubusercontent.com/Matheus-Pombeiro/canada_imigrants/refs/heads/main/assets/datasets/imigrantes_canada.csv'
```

```python
# Lendo o dataset
df = pd.read_csv(url)
df
```

```python
# Obtendo informações do dataset
df.info()
```

```python
# Alterando o index
df.set_index('País', inplace=True)
```

```python
# Definindo um lista com anos
anos = list(map(str, range(1980, 2014)))
anos
```

```python
# Obtendo os registros a respeito do Brasil
brasil = df.loc['Brasil', anos]
brasil
```

```python
# Criando um dataframe
brasil_dict = {'ano': brasil.index.tolist(), 'imigrantes': brasil.values.tolist()}
dados_brasil = pd.DataFrame(brasil_dict)
dados_brasil
```

```python
# Determinando o tamanho do gráfico
plt.figure(figsize=(8, 4))

# Plotando um gráfico sobre os imigrantes em função do tempo
plt.plot(dados_brasil['ano'], dados_brasil['imigrantes'])

# Determinando um título
plt.title('Imigração do Brasil para o Canadá')

# Determinando rótulos
plt.xlabel('Ano')
plt.ylabel('Número de Imigrantes')

# Determinando o intervalo de tempo para o eixo x
plt.xticks(['1980', '1985', '1990', '1995', '2000', '2005', '2010'])

# Determinando valores para o eixo y
plt.yticks([500, 1000, 1500, 2000, 2500, 3000])

# Exibindo o gráfico
plt.show()
```

## Criando figuras com Matplotlib

```python
# Criando uma figura
'''fig -> cria a figura (espaço em branco)
ax ->  representa os eixos (axes)'''
fig, ax = plt.subplots(figsize=(8, 4))

# Passando os dados para os eixos (x, y)
ax.plot(dados_brasil['ano'], dados_brasil['imigrantes'])

# Determinando um título
ax.set_title('Imigração do Brasil para o Canadá\nde 1980 à 2013')

# Determinando os rótulos
ax.set_xlabel('Ano')
ax.set_ylabel('Número de Imigrantes')

# Manipulando o eixo x, determinando o intervalo de tempo de 5 em 5 anos
ax.xaxis.set_major_locator(plt.MultipleLocator(5))

# Exibindo o gráfico
plt.show()
```

```python
# Criando uma figura com mais de um gráfico (axs), com 1 linha e 2 colunas
fig, axs = plt.subplots(1, 2, figsize=(15, 5))

# -------------------------------------------------------

# Passando os dados para o 1º gráfico e plotando ele
axs[0].plot(dados_brasil['ano'], dados_brasil['imigrantes'])

# Definindo o título para o 1º gráfico
axs[0].set_title('Imigração do Brasil para o Canadá\nde 1980 à 2013')

# Definindo os rótulos do 1º gráfico
axs[0].set_xlabel('Ano')
axs[0].set_ylabel('Número de Imigrantes')

# Definindo a distância entre os valores no 1º gráfico
axs[0].xaxis.set_major_locator(plt.MultipleLocator(5))

# Definindo a grade do 1º gráfico
axs[0].grid()

# -------------------------------------------------------

# Passando os dados para o 2º gráfico e plotando ele
axs[1].boxplot(dados_brasil['imigrantes'])

# Definindo o título para o 2º gráfico
axs[1].set_title('Boxplot da imigração do Brasil para o Canadá\nde 1980 à 2013')

# Definindo os rótulos do 2º gráfico
axs[1].set_xlabel('Brasil')
axs[1].set_ylabel('Número de Imigrantes')

# Definindo a grade do 1º gráfico
axs[1].grid()

# -------------------------------------------------------

# Exibindo os gráficos
plt.show()
```

```python
# Criando uma figura com mais de um gráfico (axs), com 2 linhas e 2 colunas
fig, axs = plt.subplots(2, 2, figsize=(10, 6))

# Definindo o espaçamento entre os gráficos
fig.subplots_adjust(hspace=0.5, wspace=0.3)

# Definindo um título geral
fig.suptitle('Imigração dos quatro maiores países da América do Sul para o Canadá\nde 1980 à 2013')

# -------------------------------------------------------

# Passando os dados para o 1º gráfico e plotando ele
axs[0, 0].plot(df.loc['Brasil', anos])

# Definindo o título para o 1º gráfico
axs[0, 0].set_title('Brasil')

# -------------------------------------------------------

# Passando os dados para o 2º gráfico e plotando ele
axs[0, 1].plot(df.loc['Colômbia', anos])

# Definindo o título para o 2º gráfico
axs[0, 1].set_title('Colômbia')

# -------------------------------------------------------

# Passando os dados para o 3º gráfico e plotando ele
axs[1, 0].plot(df.loc['Argentina', anos])

# Definindo o título para o 3º gráfico
axs[1, 0].set_title('Argentina')

# -------------------------------------------------------

# Passando os dados para o 4º gráfico e plotando ele
axs[1, 1].plot(df.loc['Peru', anos])

# Definindo o título para o 4º gráfico
axs[1, 1].set_title('Peru')

# -------------------------------------------------------

# Definindo os intervalos de valores
for ax in axs.flat:
  ax.xaxis.set_major_locator(plt.MultipleLocator(5))

# Definindo os rótulos
for ax in axs.flat:
  ax.set_xlabel('Ano')
  ax.set_ylabel('Número de Imigrantes')

# -------------------------------------------------------

# Definindo os valores mínimo e máximo
ymin = 0
ymax = 7000

# Implementando os valores min e max de y em cada gráfico
for ax in axs.ravel():
  ax.set_ylim(ymin, ymax)

# -------------------------------------------------------

# Exibindo os gráficos
plt.show()
```

```python
# Obtendo países da américa do sul
america_sul = df.query('Região == "América do Sul"').sort_values('Total', ascending=True)
america_sul
```

```python
# Declarando uma lista de cores para ser usada no gráfico
cores = ['royalblue', 'orange', 'forestgreen', 'orchid', 'purple', 'brown', 'slateblue', 'gray', 'olive', 'navy', 'teal', 'tomato']
```

```python
# Declarando uma lista de cores vazia para ajudar a destacar o Brasil
cores2 = []

# Iterando sobre essa lista
for pais in america_sul.index:
  if pais == 'Brasil':
    cores2.append('green')
  else:
    cores2.append('silver')
```

```python
# Criando uma figura
fig, ax = plt.subplots(figsize=(12, 5))

# Plottando um gráfico de barras
ax.barh(america_sul.index, america_sul['Total'], color=cores2)

# Costumizando o gráfico
ax.set_title('América do Sul: Brasil foi o quarto país com mais imigrantes\npara o Canadá de 1980 à 2013', loc='left', fontsize=18)
ax.set_xlabel('Número de Imigrantes', fontsize=14)
ax.set_ylabel('', fontsize=14)
ax.xaxis.set_tick_params(labelsize=12)
ax.yaxis.set_tick_params(labelsize=12)

# Adicionando uma tag em cada barra
for i, v in enumerate(america_sul['Total']):
  ax.text(v + 20, i, str(v), color='black', fontsize=10, ha='left', va='center')

# Retirando a borda do gráfico
ax.set_frame_on(False)

# Retirando o eixo x
ax.get_xaxis().set_visible(False)

# Retirando os ticks
ax.tick_params(axis='both', which='both', length=0)

# Salvando o gráfico
fig.savefig('imigracao_america_sul.png', transparent=False, dpi=300, bbox_inches='tight')

# Exibindo o gráfico
plt.show()
```

```python
# Criando uma figura para p gráfico
fig, axs = plt.subplots(2, 2, figsize=(14, 8))

# Ajustando o espaçamento entre os gráficos
plt.subplots_adjust(hspace=0.5, wspace=0.3)

# Determinando um título para o conjunto de gráficos
fig.suptitle('Vendas no período de janeiro a dezembro de 2022 nas lojas A,B,C e D', fontsize=18)

# Determinando uma lista de cores para as diferentes lojas
cores = ['darkviolet', 'green', 'darkblue', 'coral']

# Determinando um loop para plotar e costumizar os gráficos
for i, ax in enumerate(axs.flat):
  ax.plot(df_vendas.loc[df_vendas.index[i]], color=cores[i], lw=3)
  ax.set_title(f'Vendas na loja {df_vendas.index[i]}', loc='left', fontsize=16)
  ax.set_xlabel('Mês', fontsize=14)
  ax.set_ylabel('Número de Vendas', fontsize=14)
  ax.tick_params(labelsize=12)
  ax.grid(color='lightgray')
```

## Conhecendo a biblioteca Seaborn

```python
# Importando a biblioteca Seaborn
import seaborn as sns
```

```python
# Configurando o tema padrão
sns.set_style()
```

```python
# Determinando um novo dataframe
top_10 = df.sort_values('Total', ascending=False).head(10)
top_10
```

```python
# Criando uma figura para plotar o gráfico
fig, ax = plt.subplots(figsize=(8, 4))

# Plotando um gráfico de barras
ax = sns.barplot(data=top_10, x='Total', y=top_10.index, orient='h')

# Determinando o título
ax.set_title('Países com maior imigração para o Canadá\nde 1980 à 2013', loc='left', fontsize=16)

# Determinando rótulos
ax.set_xlabel('Número de Imigrantes', fontsize=14)
ax.set_ylabel('')


# Exibindo o gráfico usando a Matplotlib
plt.show()
```

```python
# Declarando uma função que irá gerar paletas de cores para os gráficos
def gerar_grafico_paleta(palette):

  # Criando a figura
  fig, ax = plt.subplots(figsize=(8, 4))

  # Plottando o gráfico
  ax = sns.barplot(data=top_10, y=top_10.index, x='Total', orient='h', palette=palette, hue=top_10.index)
  # O hue deve ser igual à variável do eixo y

  # Definindo o título
  ax.set_title('Países com maior imigração para o Canadá\nde 1980 à 2013', loc='left', fontsize=18)

  # Definindo os rótulos
  ax.set_xlabel('Número de Imigrantes', fontsize=14)
  ax.set_ylabel('')

  # Removendo as bordas (frame)
  sns.despine()

  # Exibindo o gráfico
  plt.show()
```

```python
# Chamando a função
gerar_grafico_paleta('tab10')
```

```python
# Configurando o tema
sns.set_theme(style='ticks')

# Chamando a função
gerar_grafico_paleta('tab10')
```
